### Cartoon

````python
#START CODE HERE (Do all your processing here)
cartoon_gray = cv2.cvtColor(cartoon, cv2.COLOR_BGR2GRAY)
_, mask = cv2.threshold(cartoon_gray, 240, 255, cv2.THRESH_BINARY_INV)
cartoon_y_mentee = cartoon.copy()
cartoon_y_mentee[mask == 0] = (255, 255, 255)
cartoon_y_mentee = cv2.cvtColor(cartoon_y_mentee, cv2.COLOR_BGR2RGB) #(in RGB format only)

#END CODE HERE (the final processed image should be cartoon_y_mentee)
````
